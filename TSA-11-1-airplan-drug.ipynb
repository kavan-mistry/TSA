```python
df = pd.read_csv("D:\labs\data_set\AusAntidiabeticDrug.csv",parse_dates=['ds'])
df
```

```python
df["ds"] = pd.to_datetime(df["ds"])
print('Shape :',df.shape)
df.head()
```

```python
df["year"] = df["ds"].dt.year
df["month"] = df["ds"].dt.strftime('%b')
df.head()
```

```python
plt.figure(figsize=(16,7))
sns.pointplot(x="month",y="y", hue="year", data=df,order=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel("month")
plt.ylabel("$ millions")
plt.title("Anti-Diabetic Drug Sales")
plt.legend(loc='upper right')
```

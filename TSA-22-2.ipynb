```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\labs\data_set\Marathon.csv")
df
```

```python
x=df['Year']
y=df['Minutes']
```

```python
plt.plot(x,y)
```

```python
m, c = np.polyfit(x, y, 1)
```

```python
print(f"The slope is {m:.2f}")
print(f"The intercept is {c:.2f}")
```

```python
y_pred = m * x + c
```

```python
x = np.array(df['Year'])
y = np.array(df['Minutes'])

plt.plot(x, y, color='purple')
plt.plot(x, y_pred, color='green')
plt.show()
```

```python
residuals = y - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(x, residuals, color='purple')
plt.axhline(y=0, color='green', linestyle='--', linewidth=1)
plt.title('Residuals vs Year')
plt.xlabel('Year')
plt.ylabel('Residuals')
plt.show()
```

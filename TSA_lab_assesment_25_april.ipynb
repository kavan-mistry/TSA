```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import arma_order_select_ic
```

```python
import warnings

# Suppressing ARIMA convergence warnings for this example
warnings.filterwarnings("ignore")
```

```python
# Fit ARIMA(2,1,0) model
arima_210 = ARIMA(differenced_data, order=(2,1,0))
arima_210_results = arima_210.fit()

# Fit ARIMA(0,1,1) model
arima_011 = ARIMA(differenced_data, order=(0,1,1))
arima_011_results = arima_011.fit()

# Compare AIC values
print("ARIMA(2,1,0) AIC:", arima_210_results.aic)
print("ARIMA(0,1,1) AIC:", arima_011_results.aic)
```

```python
# Extract ARIMA(2,1,0) model results
arima_210_results = arima_210.fit()

# Extract parameter estimates, standard errors, and confidence intervals
intercept = arima_210_results.params[0]
ar1_est, ar2_est = arima_210_results.arparams
ar1_std_err, ar2_std_err = arima_210_results.bse[1:3]
ar1_conf_int = arima_210_results.conf_int()[0]
ar2_conf_int = arima_210_results.conf_int()[1]

# Calculate original parameters based on differenced data
ar1_orig = -ar1_est
ar2_orig = -ar2_est
intercept_orig = -(ar1_est + ar2_est - 1) * intercept

# Print parameter estimates, standard errors, and confidence intervals
print("Parameter Estimates for AR(2) Model:")
print("-------------------------------------")
print("Parameter    Estimate    Standard Error    95% Confidence Interval")
print("--------------------------------------------------------------------")
print(f"Intercept     {intercept_orig:.4f}        {arima_210_results.bse[0]:.4f}        ")
print(f"AR1           {ar1_orig:.4f}        {ar1_std_err:.4f}      ({ar1_conf_int[0]:.4f}, {ar1_conf_int[1]:.4f})")
print(f"AR2           {ar2_orig:.4f}        {ar2_std_err:.4f}      ({ar2_conf_int[0]:.4f}, {ar2_conf_int[1]:.4f})")
print()

# Print additional information
print(f"Number of Observations:               {arima_210_results.nobs}")
print(f"Degrees of Freedom:                   {arima_210_results.nobs - 3}")
print(f"Residual Standard Deviation:          {arima_210_results.scale:.4f}")
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming arima_210 is the fitted ARIMA(2,1,0) model

# Fit the ARIMA model
arima_210_results = arima_210.fit()

# Obtain residuals
residuals = arima_210_results.resid

# Create 4-plot
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Time series plot
sm.graphics.tsa.plot_acf(residuals, lags=40, ax=axes[0, 0])
axes[0, 0].set_title('Autocorrelation of Residuals')

# Histogram of residuals
axes[0, 1].hist(residuals, bins=20, density=True)
axes[0, 1].set_title('Histogram of Residuals')

# Q-Q plot of residuals
sm.qqplot(residuals, line='s', ax=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot of Residuals')

# Scatterplot of residuals against lagged values
sm.graphics.tsa.plot_acf(residuals, lags=40, ax=axes[1, 1], zero=False)
axes[1, 1].set_title('Autocorrelation of Residuals')

plt.tight_layout()
plt.show()
```

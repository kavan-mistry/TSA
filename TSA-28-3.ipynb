# EXPONENTIAL SMOOTHING

```python
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Dow Jones'])
plt.title('Dow Jones Index Over Time')
plt.xlabel('Date')
plt.ylabel('Dow Jones Index')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## FIRST-ORDER EXPONENTIAL SMOOTHING

```python
def smooth(y, alpha=0.2):
    smoothed_y = np.zeros_like(y)
    smoothed_y[0] = y[0]

    for i in range(1, len(y)):
        smoothed_y[i] = alpha * y[i] + (1 - alpha) * smoothed_y[i - 1]

    return smoothed_y
```

```python
smoothed_y = smooth(df['Dow Jones'], alpha=0.2)
```

```python
smoothed_y2 = smooth(df['Dow Jones'], alpha=0.4)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Dow Jones'], label='Original')
plt.plot(df['Date'], smoothed_y, label='Smoothed')
plt.title('Dow Jones Index Over Time')
plt.xlabel('Date')
plt.ylabel('Dow Jones Index')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Dow Jones'], label='Original')
plt.plot(df['Date'], smoothed_y2, label='Smoothed')
plt.title('Dow Jones Index Over Time')
plt.xlabel('Date')
plt.ylabel('Dow Jones Index')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Dow Jones'], label='Original')
plt.plot(df['Date'], smoothed_y, label='Smoothed , 0.2')
plt.plot(df['Date'], smoothed_y2, label='Smoothed , 0.4')
plt.title('Dow Jones Index Over Time')
plt.xlabel('Date')
plt.ylabel('Dow Jones Index')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## SECOND-ORDER EXPONENTIAL SMOOTHING

```python
def smoothsecond(y, alpha=0.2, lambda_=0.3):
    smoothed_y = np.zeros_like(y)
    smoothed_y[0] = y[0]

    for i in range(1, len(y)):
        smoothed_y[i] = alpha * y[i] + (1 - alpha) * smoothed_y[i - 1] - lambda_ * (smoothed_y[i - 1] - smoothed_y[i - 2])

    return smoothed_second_y
```

```python
smoothed_second_y = smooth(df['Dow Jones'], alpha=0.2)
smoothed_second_y
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Dow Jones'], label='Original')
plt.plot(df['Date'], smoothed_second_y, label='Smoothed')
plt.title('Dow Jones Index Over Time (SECOND-ORDER EXPONENTIAL SMOOTHING)')
plt.xlabel('Date')
plt.ylabel('Dow Jones Index')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

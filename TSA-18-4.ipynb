Two realizations of the ARMA(1,1) model: 
* (a) yt = 16 + 0.6yt‚àí1 + ùúÄt + 0.8ùúÄt‚àí1  
* (b) yt = 16 ‚àí 0.7yt‚àí1 + ùúÄt ‚àí 0.6ùúÄt‚àí1 
* (c) The ACF of (a) 
* (d) the ACF of (b) 
* (e) the PACF of (a) 
* (f) the PACF of (b)

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
import numpy as np
```

```python
# Model (a)
model_a = ARIMA(np.random.rand(100), order=(1, 0, 1))  # Replace with your actual data
model_a.arparams = np.array([0.6])
model_a.maparams = np.array([1, 0.8])

# Model (b)
model_b = ARIMA(np.random.rand(100), order=(1, 0, 1))  # Replace with your actual data
model_b.arparams = np.array([-0.7])
model_b.maparams = np.array([1, -0.6])
```

```python
model_a_fit = model_a.fit()
model_b_fit = model_b.fit()
```

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
acf_a = acf(model_a_fit.resid, nlags=20)  # Use residuals from the fitted model
acf_b = acf(model_b_fit.resid, nlags=20)

pacf_a = pacf(model_a_fit.resid, nlags=20)
pacf_b = pacf(model_b_fit.resid, nlags=20)
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
# Plot ACF for model (a)
plt.figure(figsize=(10, 6))
plt.stem(acf_a)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)  # Add horizontal line at 0
plt.title('ACF of Model (a)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Plot ACF for model (b) (similar to model (a))
plt.figure(figsize=(10, 6))
plt.plot(acf_b)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)
plt.title('ACF of Model (b)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Plot PACF for model (a)
plt.figure(figsize=(10, 6))
plt.plot(pacf_a)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)
plt.title('PACF of Model (a)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()

# Plot PACF for model (b) (similar to model (a))
plt.figure(figsize=(10, 6))
plt.plot(pacf_b)
plt.axhline(0, color='black', linestyle='--', alpha=0.7)
plt.title('PACF of Model (b)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Define the parameters for the ARMA(1,1) models
params_a = np.array([0.6, 0.8])
params_b = np.array([-0.7, -0.6])

# Generate artificial data for visualization
np.random.seed(0)
y_a = sm.tsa.arma_generate_sample(ar=params_a, ma=[1], nsample=1000)
y_b = sm.tsa.arma_generate_sample(ar=params_b, ma=[1], nsample=1000)

# Plot the ACF and PACF for both models
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Model (a) plots
sm.graphics.tsa.plot_acf(y_a, lags=30, ax=axs[0, 0], title='ACF of Model (a)')
sm.graphics.tsa.plot_pacf(y_a, lags=30, ax=axs[0, 1], title='PACF of Model (a)')

# Model (b) plots
sm.graphics.tsa.plot_acf(y_b, lags=30, ax=axs[1, 0], title='ACF of Model (b)')
sm.graphics.tsa.plot_pacf(y_b, lags=30, ax=axs[1, 1], title='PACF of Model (b)')

plt.tight_layout()
plt.show()
```
